```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
"genre": [
    "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
    "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
    "Pop","Rock","Hip-Hop","Electronic"
],
"streams": [
    120,90,150,70,130,95,140,80,
    110,85,155,75,125,100,145,78,
    118,92,160,82
],
"likes": [
    30,20,50,15,35,22,48,18,
    28,19,55,16,32,25,47,17,
    29,21,60,19
],
"shares": [
    10,8,20,5,12,9,18,6,
    9,7,22,5,11,10,19,6,
    10,8,25,7
],
"year": [
    2020,2020,2020,2020,2021,2021,2021,2021,
    2022,2022,2022,2022,2023,2023,2023,2023,
    2024,2024,2024,2024
]
}

df = pd.DataFrame(data)
```

```python
# Exercise 1

total_streams_by_genre = df.groupby('genre')['streams'].sum()

plt.figure(figsize=(9, 6))

plt.bar(total_streams_by_genre.index, total_streams_by_genre, color="midnightblue")

plt.title("Total Steams By Genre", fontsize=18, pad=18)
plt.xlabel("Genre")
plt.ylabel("Total Streams")

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.tight_layout()
plt.show()
```

```python
# Exercise 2

likes = df['likes']
streams = df['streams']
plt.figure(figsize=(9, 6))

plt.scatter(likes, streams, color="forestGreen", alpha=0.9, s=80)

plt.title("Do more steams = more likes?", fontsize=18, pad=18)
plt.xlabel("Streams")
plt.ylabel("Likes")

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Exercise 3

mariana_trench_blue = "#00192c"


plt.figure(figsize=(9,6))

plt.hist(df['streams'], bins=8, color=mariana_trench_blue, edgecolor="white")

plt.title('Distribution of Streams', fontsize=18, pad=18)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```
