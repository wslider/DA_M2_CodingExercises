```python
import pandas as pd
import numpy as np

data = {
"Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
"Age": [25, np.nan, 35, 40, None],
"City": ["NYC", "LA", None, "Miami", "Boston"],
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
```

```python
df['City'] = df['City'].fillna("Unknown")
print(df)
```
